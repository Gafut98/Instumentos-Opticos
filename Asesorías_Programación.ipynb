```python
nm = 1e-9
um = 1e-6
mm = 1e-3

#wave_Length = 650*1e-9
wave_length = 650*nm
'''Voy a crear un arreglo de 0 a 5u en pasos de 10nm
Sobre eso definimos la priera onda en una dimensión
'''
#Primero definimos los parámetros de la onda
wave_range1 = np.arange(0,5*um,10*nm)
k = 2*np.pi/wave_length


#Una vez definidos los parámetros de la onda, definimos ahora si la onda
onda1D=np.sin(k*wave_range1)

#Ahora vemos la onda en una gráfica
plt.plot(onda1D)
plt.show()
```

```python
#otra forma de realizar el ejercicio 
#anterior es con intervalos, en vez de listas, de modo que usamos linspace
wave_length = 650*nm
wave_range = np.linspace(0,5*um,1000)
k = 2*np.pi/wave_length
onda1D = np.sin(k*wave_range)

plt.plot(onda1D)
plt.show()
```

```python
plt.plot(np.abs(onda1D))
plt.show()
```

```python
FFT_onda1D = np.fft.fft(onda1D)
```

```python

plt.plot(
    np.abs(
        FFT_onda1D
    )**2
)
```

```python
#Notemos que la gráfica no correaponde a lo que se suponde debería ser la TF
# debemos entonces usar el comando shift para reubicar los componentes frecuenciales
plt.plot(np.abs(np.fft.fftshift(FFT_onda1D))**2)
```

```python
#Para indexar, es decir tomar un rango diferentede lo anterior usamos[A:B]
#Con A y B los nuevos rangos disponibes pertenceientres al arreglo
plt.plot((np.abs(np.fft.fftshift(FFT_onda1D))**2)[450:550])
```
